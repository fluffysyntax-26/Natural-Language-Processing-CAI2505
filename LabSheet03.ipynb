# Lab Sheet 3: Sentiment Analysis and Named Entity Recognition

<b>Aim: </b>To perform Sentiment Analysis

```python
from textblob import TextBlob
```

```python
sentences = [ 
    "The movie was fantastic", 
    "The acting was excellent",
    "I would not recommend this film", 
    "Not a good movie",
    "Very bad movie",
    "The movie was average", 
    "Not Good, Not Bad"
]
```

```python
polarities = []
```

```python
for sentence in sentences: 
    blob = TextBlob(sentence)
    print(blob)
    polarity = blob.sentiment.polarity
    polarities.append(polarity)

    if polarity > 0: 
        sentiment = "Postive"
    elif polarity < 0: 
        sentiment = "Negative"
    else: 
        sentiment = "Neutral"

    print(f"Polarity: {polarity}, Sentiment: {sentiment}\n")
```

## Named Entity Recognition

```python
import spacy 
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Elon Musk is the CEO of Tesla and SpaceX based in the United States"

doc = nlp(text)

for ent in doc.ents: 
    print(ent.text, "->", ent.label_)
```

```python
# Entity Frequency Analysis
entity_count = {}

for ent in doc.ents: 
    entity_count[ent.label_] = entity_count.get(ent.label_, 0) + 1

print("Entity Count: ", entity_count)
```
